```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train2 = pd.DataFrame(sc_X.fit_transform(X_train))
X_test2 = pd.DataFrame(sc_X.transform(X_test))
X_train2.columns = X_train.columns.values
X_test2.columns = X_test.columns.values
X_train2.index = X_train.index.values
X_test2.index = X_test.index.values
X_train = X_train2
X_test = X_test2
```

```python
from catboost import CatBoostClassifier
# classifier = XGBClassifier(learning_rate =0.1, n_estimators=210, max_depth=3,min_child_weight=3,scale_pos_weight=1)
classifier = CatBoostClassifier(iterations=210,learning_rate =0.1,
                           task_type="GPU",loss_function='CrossEntropy',eval_metric='AUC',
                           random_seed=42)
classifier.fit(X_train, y_train)
```

```python
from xgboost import XGBClassifier
classifier2 = XGBClassifier(learning_rate =0.1, n_estimators=210, max_depth=7,scale_pos_weight=1)
classifier2.fit(X_train, y_train)
```

```python
import lightgbm as lgb
classifier3 = lgb.LGBMClassifier(boosting_type='dart',learning_rate=0.18, max_depth=7,
               n_estimators=450,objective='binary',
               random_state=42)
classifier3.fit(X_train, y_train)
```

```python
 from sklearn.ensemble import RandomForestClassifier
 classifier4 =RandomForestClassifier(bootstrap=True, ccp_alpha=0.0, class_weight=None,
                       criterion='entropy', max_depth=7, max_features='auto',
                       max_leaf_nodes=3, max_samples=None,
                       min_impurity_decrease=0.0,
                       min_samples_leaf=1, min_samples_split=3,
                       min_weight_fraction_leaf=0.0, n_estimators=400,
                       n_jobs=-1, oob_score=False, random_state=8681, verbose=0,
                       warm_start=False)
 classifier4.fit(X_train, y_train)
```

```python
# Predicting Test Set
y_pred = classifier.predict(X_test)
y_pred2 = classifier2.predict(X_test)
y_pred3 = classifier3.predict(X_test)
y_pred4 = classifier4.predict(X_test)
```

```python
# Evaluating Results
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
cm = confusion_matrix(y_test, y_pred)
accuracy_score(y_test, y_pred)
precision_score(y_test, y_pred) # tp / (tp + fp)
recall_score(y_test, y_pred) # tp / (tp + fn)
print(f1_score(y_test, y_pred,average='macro'))
print(f1_score(y_test, y_pred2,average='macro'))
print(f1_score(y_test, y_pred3,average='macro'))
print(f1_score(y_test, y_pred4,average='macro'))

```
